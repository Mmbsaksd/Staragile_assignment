```python
for url in urls:
    page = requests.get(url)
    if page.status_code !=200:
        print(f"Failed to fetch content from :{url}. Status Code: {page.status_code}")
else:
    print("Successfully loaded all url content")
```

```python
stocks = []
for url in urls:
    stock = {}
    try:
        page = requests.get(url)
        
        soup = BeautifulSoup(page.text,'html.parser')
        company = soup.find('h1',{'class':"margin-0 show-from-tablet-landscape"})
        stock["Company"] = company.text if company else np.nan
        columns = ['Market Cap','Current Price','High','Low','Stock P/E','Book Value','Dividend Yield','ROCE','ROE','Face Value']
        figure = soup.find_all('span',{'class':"number"})
        for i in range(len(columns)):
            stock[columns[i]] = figure[i].text if figure[i] else np.nan
            
            
        sector = soup.find_all('a', href = lambda herf:herf and herf.startswith("/company/compare/"))
        stock["Sector"] = sector[0].text.strip() if sector[0] else np.nan
        stock["Industry"] = sector[1].text.strip() if sector[1] else np.nan
        
        stocks.append(stock)
        
    except Exception as e:
        print(f'Error scraping data from {url}:{e}')
        
stocks
```

```python
df = pd.DataFrame(stocks)
df.head(30)
```

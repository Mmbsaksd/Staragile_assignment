```python
#Checking for executions
for url in urls:
    page = requests.get(url)
    if page.status_code !=200:
        print(f"Failed to fetch content from :{url}. Status Code: {page.status_code}")
else:
    print("Successfully loaded all url content")
```

```python
#Scrapping and collecting data in dict
stocks = []
for url in urls:
    stock = {}
    try:
        page = requests.get(url)

        soup = BeautifulSoup(page.text,'html.parser')
        company = soup.find('h1',{'class':"margin-0 show-from-tablet-landscape"})
        stock["Company"] = company.text if company else np.nan
        columns = ['Market Cap','Current Price','High','Low','Stock P/E','Book Value','Dividend Yield','ROCE','ROE','Face Value']
        figure = soup.find_all('span',{'class':"number"})
        for i in range(len(columns)):
            stock[columns[i]] = figure[i].text if figure[i] else np.nan


        sector = soup.find_all('a', href = lambda herf:herf and herf.startswith("/company/compare/"))
        stock["Sector"] = sector[0].text.strip() if sector[0] else np.nan
        stock["Industry"] = sector[1].text.strip() if sector[1] else np.nan

        stocks.append(stock)

    except Exception as e:
        print(f'Error scraping data from {url}:{e}')

stocks
```

```python
#Converting to DataFrame
df = pd.DataFrame(stocks)
df.head(30)
```

# Data Analysis

```python
#Loading required libraries for Data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

# Set the seaborn style
style.use('seaborn')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('stocks.csv')
df.head(10)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['Stock P/E'].replace(np.nan,df['Stock P/E'].mean(),inplace=True)
df.ROE.replace(np.nan,df.ROE.mean(),inplace=True)
df.isnull().sum()
```

```python
df.Sector.value_counts()
```

```python
sector = ['Pharmaceuticals','Banks','FMCG','IT - Software','Finance','Automobile','Refineries','Power Generation & Distribution','Insurance','Cement','Mining & Mineral products','Gas Distribution']
def sector(x):
    sector = ['Pharmaceuticals','Banks','FMCG','IT - Software','Finance','Automobile','Refineries','Power Generation & Distribution','Insurance','Cement','Mining & Mineral products','Gas Distribution']
    if x in sector:
        return x
    else:
        return 'Other'
```

```python
df['Sector'] = df['Sector'].apply(sector)
df.Sector.value_counts()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.applymap(lambda x: str(x).replace(',',''))
numeric_colm = ['Market Cap', 'Current Price', 'High', 'Low', 'Stock P/E', 'Book Value', 'Dividend Yield', 'ROCE', 'ROE', 'Face Value']
df[numeric_colm] = df[numeric_colm].apply(pd.to_numeric,errors='coerce')
df.info()
```

```python
# Plotting Market Cap distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Market Cap'], bins=20, kde=True, color='blue')
plt.title('Distribution of Market Cap')
plt.xlabel('Market Cap')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Stock P/E distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Stock P/E'], bins=20, kde=True, color='green')
plt.title('Distribution of Stock P/E')
plt.xlabel('Stock P/E')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Dividend Yield distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Dividend Yield'], bins=20, kde=True, color='orange')
plt.title('Distribution of Dividend Yield')
plt.xlabel('Dividend Yield')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plotting ROCE vs ROE
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ROE', y='ROCE', data=df, color='red')
plt.title('ROE vs ROCE')
plt.xlabel('ROE')
plt.ylabel('ROCE')
plt.show()
```

```python
# Plotting Sector-wise Market Cap
plt.figure(figsize=(10, 6))
sns.barplot(x='Sector', y='Market Cap', data=df, ci=None)
plt.title('Sector-wise Market Cap')
plt.xlabel('Sector')
plt.ylabel('Market Cap')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting Industry-wise Market Cap
plt.figure(figsize=(10, 6))
sns.barplot(x='Industry', y='Market Cap', data=df, ci=None)
plt.title('Industry-wise Market Cap')
plt.xlabel('Industry')
plt.ylabel('Market Cap')
plt.xticks(rotation=90)
plt.show()
```

 ### **Summery of project :**
 


1. **Market Capitalization**:
   - Most companies have market capitalizations ranging from 1 lakh crore to 20 lakh crore.
   
2. **P/E Ratio**:
   - The majority of companies have P/E ratios around 40 to 50, with some exceptions reaching as high as 300.

3. **Dividend Yield**:
   - Many companies either do not provide any dividend yield or offer yields less than 3%. However, there are rare companies in the Nifty 100 index that offer dividend yields as high as 30%.

4. **Relationship between ROCE and ROE**:
   - There is a positive relationship between Return on Capital Employed (ROCE) and Return on Equity (ROE). This suggests that companies with higher ROCE tend to have higher ROE.

5. **Sector-wise Market Capitalization**:
   - Refineries have the highest market capitalization among the sectors represented in the Nifty 100 index, while gas distribution has the lowest.

6. **Distribution of Market Capitalization across Industries**:
   - The telecommunications industry, particularly telecommunications services, holds the largest share of market capitalization in the Nifty 100 index. Conversely, the hotel industry has a smaller share.

These insights provide valuable information about the composition and characteristics of companies listed in the Nifty 100 index. They can be used for further analysis and investment decision-making.

## 1.	What is probability? List down some laws of probability?

**Probability:**

Probability is a branch of mathematics that quantifies the likelihood of various outcomes in a given event. It is expressed as a number between 0 and 1, where 0 indicates impossibility, 1 indicates certainty, and values between 0 and 1 represent the likelihood of an event occurring.

**Laws of Probability:**

1. **Law of Addition (Sum Rule):**

- What it does: Calculates the probability of event A or event B or both occurring.
- Formula: P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
- Example: Rolling a die and wanting either a 1 or a 6. P(1 or 6) = P(1) + P(6) - P(both 1 and 6) = (1/6) + (1/6) - 0 = 1/3

2. **Law of Multiplication (Product Rule):**

- What it does: Calculates the probability of both event A and event B occurring together.
- Formula: P(A ∩ B) = P(A|B) * P(B)
- Example: Drawing two black cards from a deck without replacement. P(2 black cards) = P(black | black drawn first) * P(black) = (3/51) * (12/52) = 3/21

3. **Complement Rule:**

- What it does: Calculates the probability of the opposite of event A happening.
- Formula: P(A') = 1 - P(A)
- Example: Knowing the sun has a 70% chance of shining (P(sun) = 0.7). P(no sun) = 1 - P(sun) = 1 - 0.7 = 0.3
4. **Law of Total Probability:**

- What it does: Calculates the probability of event B occurring by considering all possible ways it could happen through different partition events.
- Formula: P(B) = Σ P(A_i ∩ B)
- Example: Three boxes with coins, each with different possibilities of heads. P(heads) = P(heads from box 1) + P(heads from box 2) + P(heads from box 3)

## 2. What is the difference between independent and dependent events?

**Independent events:**

No influence: These are events where the outcome of one event has no bearing on the likelihood of another event occurring.
No matter what: Whether the first event happens or not, the probability of the second event remains the same.
Examples: Flipping a coin twice (each flip is independent of the other), drawing a card from a deck without replacing it (the remaining cards don't change the probability of the next draw).

**Dependent events:**

Influence each other: The outcome of one event affects the probability of the second event occurring.
It matters: Knowing the outcome of the first event gives you more information about the likelihood of the second event happening.
Examples: Drawing a card from a deck and then drawing another without replacing it (the second draw has fewer cards), throwing a dart at a board and then throwing another (knowing the location of the first dart can guide the second).


## 3.	What is conditional probability? What is the formula to calculate conditional probability?

**Conditional Probability:**

Conditional probability is the probability of an event occurring given that another event has already occurred. It represents the likelihood of one event happening under the condition that another event has taken place.

**Formula for Conditional Probability:**

The formula for conditional probability is expressed as:

**Formula:** The essential formula is P(A|B) = P(A ∩ B) / P(B).

**Interpretation:**
 - P(A|B): Conditional probability of A given B.
 - P(A ∩ B): Probability of both A and B occurring.
 - P(B): Probability of B occurring independently.
In words, the conditional probability of A given B is the probability of both A and B occurring divided by the probability of B occurring.

This formula is derived from the definition of conditional probability, emphasizing the relationship between the probability of the intersection of two events and the probability of the condition.

## 7.	How does Probability Mass Function and Probability Density Function differ?

Probability mass functions (PMFs) are used for discrete random variables, while probability density functions (PDFs) are used for continuous random variables. 



**Probability Mass Function (PMF) vs. Probability Density Function (PDF):**

1. **PMF (Probability Mass Function):**
   - **For Discrete Random Variables:** Models probabilities of distinct values.
   - **Notation:** \(P(X = x)\).
   - **Interpretation:** Direct probability values.
   - **Sum Rule:** Total probabilities sum to 1.

2. **PDF (Probability Density Function):**
   - **For Continuous Random Variables:** Models probabilities of a range of values.
   - **Notation:** \(f(x)\).
   - **Interpretation:** Gives density; integrate for probability.
   - **Integral Rule:** Total area under the curve is 1.

## 9.	What is the probability of getting a sum less than 8 when two dice are thrown?

**9. Probability of Getting a Sum Less Than 8 when Two Dice are Thrown:**

1. **Total Outcomes:**
   - Each die has 6 faces, so there are 6 x 6 = 36 possible combinations when rolling two dice.
   
2. **Desired Outcomes (Getting a Sum Less Than 8):**
   - The possible sums less than 8 are 2, 3, 4, 5, 6, and 7.

3. **Number of Desired Outcomes:**
   - For each possible sum, count the number of ways to achieve it:
     - Sum of 2: (1, 1)
     - Sum of 3: (1, 2), (2, 1)
     - Sum of 4: (1, 3), (2, 2), (3, 1)
     - Sum of 5: (1, 4), (2, 3), (3, 2), (4, 1)
     - Sum of 6: (1, 5), (2, 4), (3, 3), (4, 2), (5, 1)
     - Sum of 7: (1, 6), (2, 5), (3, 4), (4, 3), (5, 2), (6, 1)
   - The problem correctly lists the specific combinations for each possible sum less than 8, summing up to a total of 21 desired outcomes.
   
4. **Probability of Getting a Sum Less Than 8:**
   - Probability = Number of Desired Outcomes / Total Possible Outcomes
   - Probability of a sum less than 8 = 21 / 36 = 7/12  

## 10.	If there are 12 horses in a race, numbered 1 to 12, what is the probability that horses bearing number 3, 5 or 8 will win the race?

**Total outcomes:** There are 12 horses, so there are 12 possible race winners.

**Desired outcomes:** There are 3 horses you're interested in (3, 5, and 8).

**Probability:** Since any of the 3 horses can win and each outcome is equally likely, the probability is simply the number of desired outcomes divided by the total number of outcomes:

**P(Horse 3, 5, or 8 wins)** = 3 horses / 12 horses = 1/4 or 25%
